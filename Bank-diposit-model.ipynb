```python
df.head()
```

### Let's convert 'marital' , 'default' , 'housing', 'loan', 'contact' to Bainary Numbers

#### Default / housing / loan
- yes = 1
- no = 0

```python
mp1={'yes': 1 ,'no':0,}
df['default']=df['default'].map(mp1)
df['housing']=df['housing'].map(mp1)
df['loan']=df['loan'].map(mp1)
```

```python
df.head()
```

#### Marital
- Married = 1
- singale/ divorced = 0

```python
mp2={'married': 1 ,'singale':0,'divorced':0}
df['marital']=df['marital'].map(mp2)
```

```python
df.head()
```

#### Contact
- Telephone = 1
- cellular = 0

```python
mp3={'telephone': 1 ,'cellular':0}
df['contact']=df['contact'].map(mp3)
```

```python
df.head()
```

#### Poutcome
- success = 1
- nonexistent/ failure = 0

```python
mp3={'success': 1 ,'nonexistent':0,'failure':0}
df['poutcome']=df['poutcome'].map(mp3)
```
